```python
import numpy as np
import cv2
import os
import zipfile
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, Flatten, Dense, Reshape, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import dlib
import io
import bz2
import shutil
from PIL import Image as PILImage
from IPython.display import Image, display
from google.colab import files
from facenet_pytorch import MTCNN
from google.colab.patches import cv2_imshow
import shutil
mtcnn = MTCNN(keep_all=True)

def unzip_data(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"Extracted files: {os.listdir(extract_to)}")


def load_single_image(img_path, img_size=(128, 128)):
    img = cv2.imread(img_path)
    if img is not None:
        img = cv2.resize(img, img_size)
        return img
    else:
        raise ValueError(f"Failed to load image: {img_path}")

def apply_mask(image):
    masked_image = image.copy()
    h, w, _ = image.shape
    mask_height = h // 2
    masked_image[mask_height:, :] = 0
    return masked_image

def plot_reconstruction(masked, reconstructed):
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(masked)
    plt.title("Masked Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(reconstructed)
    plt.title("Reconstructed Image")
    plt.axis('off')

    plt.show()

def ssim_loss(y_true, y_pred):
    return 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))

compressed_file = "/content/shape_predictor_68_face_landmarks.dat.bz2"
decompressed_file = "/content/shape_predictor_68_face_landmarks.dat"

if os.path.exists(compressed_file):
    with bz2.BZ2File(compressed_file, 'rb') as file, open(decompressed_file, 'wb') as decompressed:
        shutil.copyfileobj(file, decompressed)
    print("Decompression complete.")

predictor_path = decompressed_file
if not os.path.exists(predictor_path):
    print("Error: shape_predictor_68_face_landmarks.dat file not found.")
else:

    zip_path = "/content/compress.zip.zip"
    extract_dir = '/content/extracted_images/'

    unzip_data(zip_path, extract_dir)

    print(f"Extracted files: {os.listdir(extract_dir)}")

    uploaded_image = files.upload()
    for file_name in uploaded_image.keys():
        image_path = file_name

        original_image = load_single_image(image_path)
        print(f"Image {file_name} uploaded successfully.")

        plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

        copied_image_path = f"/content/copied_{file_name}"
        cv2.imwrite(copied_image_path, original_image)
        copied_image = load_single_image(copied_image_path)

        rgb_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
        boxes, probs, landmarks = mtcnn.detect(rgb_image, landmarks=True)

        if boxes is not None and len(boxes) > 0:
            print(f"Number of faces detected: {len(boxes)}")

            for i, box in enumerate(boxes):

                x1, y1, x2, y2 = map(int, box)
                cv2.rectangle(original_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

            cv2_imshow(original_image)

            copied_image = copied_image / 255.0

            masked_image = apply_mask(copied_image)

            input_shape = (128, 128, 3)


            inputs = Input(shape=input_shape)
            x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
            x = Dropout(0.3)(x)
            x = Conv2D(64, (3, 3), activation='relu', padding='same', strides=2)(x)
            x = Dropout(0.3)(x)
            x = Conv2D(128, (3, 3), activation='relu', padding='same', strides=2)(x)
            x = Dropout(0.3)(x)
            x = Flatten()(x)
            latent = Dense(512, activation='relu')(x)


            x = Dense(32 * 32 * 128, activation='relu')(latent)
            x = Reshape((32, 32, 128))(x)
            x = Conv2DTranspose(128, (3, 3), activation='relu', padding='same', strides=2)(x)
            x = Conv2DTranspose(64, (3, 3), activation='relu', padding='same', strides=2)(x)
            x = Conv2DTranspose(32, (3, 3), activation='relu', padding='same')(x)
            outputs = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

            autoencoder = Model(inputs, outputs)

            autoencoder.compile(optimizer=Adam(learning_rate=1e-4), loss=ssim_loss)

            autoencoder.fit(np.expand_dims(copied_image, axis=0), np.expand_dims(copied_image, axis=0), epochs=50, batch_size=1)

            masked_image_input = np.expand_dims(masked_image, axis=0)

            reconstructed_image = autoencoder.predict(masked_image_input)

            reconstructed_image = np.clip(reconstructed_image[0] * 255.0, 0, 255).astype(np.uint8)

            plot_reconstruction(masked_image, reconstructed_image)
        else:
            print("No face detected in the uploaded image.")
```
